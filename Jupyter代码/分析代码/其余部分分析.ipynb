### 白嫖程度对比

```python
tran=pd.concat([dataset.点赞数/dataset.播放量,
           dataset.投币数/dataset.播放量,
           dataset.转发数/dataset.播放量,
           dataset.收藏数/dataset.播放量,
           dataset.弹幕数/dataset.播放量,
           dataset.评论数/dataset.播放量],
           axis=1)
tran=tran.rename(columns={0:'点赞数转化率/万',1:'投币数转化率/万',2:'转发数转化率/万',3:'收藏数转化率/万',4:'弹幕数转化率/万',5:'评论数转化率/万'})
tran=tran*10000
tran
```

```python
dataset_with_tran=pd.concat(([dataset,tran]),axis=1)
dataset_with_tran.head()
```

```python
table_transfer1=dataset_with_tran.pivot_table(values=['点赞数转化率/万', '投币数转化率/万','转发数转化率/万','收藏数转化率/万'],index='分区',aggfunc='mean')
table_transfer1=table_transfer1.reindex(columns=['点赞数转化率/万','收藏数转化率/万', '投币数转化率/万','转发数转化率/万'])
table_transfer1
```

```python
pic11=Timeline()
for cate in table_transfer1.columns:
    table_transfer1.sort_values(by=cate,ascending=True,inplace=True)
    bar=Bar()
    bar.add_xaxis(table_transfer1.index.tolist())
    bar.add_yaxis('',[round(float(value)) for value in table_transfer1.loc[:,cate].values],color='#6A93B0')
    bar.reversal_axis()
    bar.set_series_opts(label_opts = opts.LabelOpts(is_show = True,position="right"))
    bar.set_global_opts(title_opts=opts.TitleOpts('白嫖程度',title_textstyle_opts=opts.TextStyleOpts(font_size=25)),
                    #yaxis_opts=opts.AxisOpts(type_="value",axistick_opts=opts.AxisTickOpts(is_show=True),splitline_opts=opts.SplitLineOpts(is_show=True)),
                    #xaxis_opts=opts.AxisOpts(type_="category"),
                    legend_opts=opts.LegendOpts(orient='vertical',pos_right=True))
    pic11.add(bar,"%s"%cate)

pic11.render_notebook()
```
