## 预处理

```python
#播放量转换
play=[]
for i in dataset.播放量.values:
    if i[-1]=='万':
        play.append(float(i[:-1])*10000)
    else:
        play.append(int(i))

dataset.drop(columns='播放量',inplace=True)
dataset['播放量']=play
```

```python
#点赞数转换

danmu=[]
for i in dataset.点赞数.values:
    if str(i)[-1]=='万':
        danmu.append(float(i[:-1])*10000)
    else:
        danmu.append(int(i))

dataset.drop(columns='点赞数',inplace=True)
dataset['点赞数']=danmu
```

```python
#投币数转换

coin=[]
for i in dataset.投币数.values:
    if str(i)[-1]=='万':
        coin.append(float(i[:-1])*10000)
    else:
        coin.append(int(i))

dataset.drop(columns='投币数',inplace=True)
dataset['投币数']=coin
```

```python
#转发数转换

forward=[]
for i in dataset.转发数.values:
    if str(i)[-1]=='万':
        forward.append(float(i[:-1])*10000)
    else:
        forward.append(int(i))

dataset.drop(columns='转发数',inplace=True)
dataset['转发数']=forward
```

```python
#评论数转换

comment=[]
for i in dataset.评论数.values:
    if i=='NaN':
        comment.append(0)
    elif i=='':
        comment.append(0)
    else:
        comment.append(int(i))
    

dataset.drop(columns='评论数',inplace=True)
dataset['评论数']=comment
```

```python
#弹幕数转换

bullet=[]
for i in dataset.弹幕数.values:
    if str(i)[-1]=='万':
        bullet.append(float(i[:-1])*10000)
    else:
        bullet.append(int(i))

dataset.drop(columns='弹幕数',inplace=True)
dataset['弹幕数']=bullet
```

```python
#收藏数转换

collect=[]
for i in dataset.收藏数.values:
    if str(i)[-1]=='万':
        collect.append(float(i[:-1])*10000)
    else:
        collect.append(int(i))

dataset.drop(columns='收藏数',inplace=True)
dataset['收藏数']=collect
```

```python
#粉丝数转换

fan=[]
for i in dataset.粉丝数.values:
    if str(i)[-1]=='万':
        fan.append(float(i[:-1])*10000)
    else:
        fan.append(int(i))

dataset.drop(columns='粉丝数',inplace=True)
dataset['粉丝数']=fan
```

```python
#投稿数转换

post=[]
for i in dataset.投稿数.values:
    if str(i)[-1]=='万':
        post.append(float(i[:-1])*10000)
    else:
        post.append(int(i))

dataset.drop(columns='投稿数',inplace=True)
dataset['投稿数']=post
```

```python
#算出每个视频投稿是在星期几

date=[]
for i in range(dataset.shape[0]):
    date.append(pd.to_datetime(dataset.发布日期[i]).strftime('%A'))
dataset['星期几']=date
```

```python
#算出每个视频投稿是在几点

moment=[]
for i in range(dataset.shape[0]):
    moment.append(dataset.发布日期[i][-8:-6])
dataset['发布时刻']=moment
```
