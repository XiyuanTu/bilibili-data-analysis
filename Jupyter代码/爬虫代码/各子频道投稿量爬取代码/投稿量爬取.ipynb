```python
#频道索引
channel_indexs=[1,3,5,7,8,9,10,11,12,13,14,18]
```

```python
for channel_index in channel_indexs:
    if channel_index==18:       #美食区必须从主站点进去，所以要特殊处理
        driver.find_element_by_xpath('//*[@id="internationalHeader"]/div[1]/div/div[1]/ul/li[1]/a').click()
        WebDriverWait(driver,60).until(EC.presence_of_element_located((By.XPATH,'//*[@id="primaryChannelMenu"]/span[18]/div/a/span')))
        
    channel=driver.find_element_by_xpath('//*[@id="primaryChannelMenu"]/span[%d]/div/a/span'%channel_index).text.strip('0123456789+')
    driver.find_element_by_xpath('//*[@id="primaryChannelMenu"]/span[%d]/div/a/span'%channel_index).click() #进入主频道
    WebDriverWait(driver,60).until(EC.presence_of_element_located((By.XPATH,'//*[@id="subnav"]/ul/li[2]/a'))) #等第一个子频道名出现时再继续
    
    sub_channel_index=2
    while True:
        try:
            sub_channel=driver.find_element_by_xpath('//*[@id="subnav"]/ul/li[%d]/a'%sub_channel_index).text
            driver.find_element_by_xpath('//*[@id="subnav"]/ul/li[%d]/a'%sub_channel_index).click()  #选择子频道
            driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
            WebDriverWait(driver,60).until(EC.presence_of_element_located((By.XPATH,'//*[@id="videolist_box"]/div[1]/div[1]/ul/a[2]/li'))) #等“按热度排序”选项出现时再继续
            driver.find_element_by_xpath('//*[@id="videolist_box"]/div[1]/div[1]/ul/a[2]/li').click() #让视频按热度排序
            try:
                driver.find_element_by_xpath('//*[@id="videolist_box"]/div[1]/div[3]/div[2]/div/span')
            except:
                driver.find_element_by_xpath('//*[@id="videolist_box"]/div[1]/div[1]/ul/a[2]/li').click()  #有时候点了“让视频按热度排序”没效果，所以再点一次

            for month in range(1,11):
                driver.find_element_by_xpath('//*[@id="videolist_box"]/div[1]/div[3]/div[2]/div/span').click()
                period=driver.find_element_by_xpath('//*[@id="videolist_box"]/div[1]/div[3]/div[2]/div[2]/div[2]/div/div[2]/a[%d]'%month).text
                driver.find_element_by_xpath('//*[@id="videolist_box"]/div[1]/div[3]/div[2]/div[2]/div[2]/div/div[2]/a[%d]'%month).click()
                WebDriverWait(driver,60).until(EC.presence_of_element_located((By.XPATH,'//*[@id="videolist_box"]/div[2]/ul/li[1]/div/div[2]/div[3]/a'))) 
                driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
                video_num=int(re.findall(r'\d+',driver.find_element_by_xpath('//*[@id="videolist_box"]/div[2]/div[2]/div/div/span').text.split('/')[1])[0])

                data=pd.DataFrame([[channel,sub_channel,period,video_num]],columns=columns)
                video_total_num=video_total_num.append(data)

            sub_channel_index+=1
            
        except:
            break
```
